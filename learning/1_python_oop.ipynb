```python
class BankAccount:
    # Class attribute — shared by ALL instances
    bank_name = "PyBank"

    def __init__(self, owner: str, balance: float = 0.0):
        # Instance attributes — unique to each object
        self.owner   = owner
        self.balance = balance
        self._txns   = []          # _ prefix = "private by convention"

    def deposit(self, amount: float) -> None:
        if amount <= 0:
            raise ValueError("Amount must be positive")
        self.balance += amount
        self._txns.append(("deposit", amount))

    def withdraw(self, amount: float) -> None:
        if amount > self.balance:
            raise ValueError("Insufficient funds")
        self.balance -= amount
        self._txns.append(("withdraw", amount))

# Usage
acc = BankAccount("Arjun", balance=5000)
acc.deposit(1000)
print(acc.owner, acc.balance) 
print(acc.bank_name)           
```

```python
class BankAccount:

    def __init__(self, account_number, balance):
        self.__account_number=account_number
        self.__balance=balance

    def check_acc_num(self):
        return self.__account_number

    def deposit(self, amount):
        self.__balance += amount
        print(f"- total balance - {self.__balance}")

    def withdraw(self, amount):
        if amount>self.__balance:
            print("- Insufficiend funds to withdraw !!!")
        else:
            self.__balance -= amount
            print("- withdraw successful.")
            print(f"- remaining balance - {self.__balance}")

    def check_balance(self):
        return self.__balance

B1=BankAccount(1234,10_000)

B1.deposit(1_000)
B1.withdraw(89_000)
B1.withdraw(150)
```

```python
class Phone:
    def __init__(self, model):
        self.model = model

    def call_contact(self, name):
        # The "Abstraction": The user just sees this method name.
        # They don't see the complex logic below.
        print(f"Connecting to network towers...")
        print(f"Establishing secure line to {name}...")
        print(f"Ringing...")

    def take_picture(self):
        # We abstract away the aperture settings, sensor data, and file compression.
        print(f"Adjusting focus and exposure for {self.model} camera...")
        print("Click! Picture saved to gallery.")

# Usage
my_phone = Phone("Pixel 9")

# The user-friendly interface:
my_phone.call_contact("Alex")
my_phone.take_picture()
```

```python
class shape():
    def calculate_area(self):
        pass

class Circle(shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        print(f"Area of Circle is {(22/7)*self.radius*self.radius}")

class Rectangle(shape):
    def __init__(self, lenght, breadth):
        self.lenght=lenght
        self.breadth=breadth

    def calculate_area(self):
        print(f"Area of Rectangle is {self.lenght*self.breadth}")

c=Circle(7)
r=Rectangle(14,15)

c.calculate_area()
r.calculate_area()
```

```python
# Calculator default x4
class calculator:
    def __init__(self, a, b, c=None):
        self.a = a
        self.b = b
        self.c = c

    def multiply(self):
        if self.c is not None:
            print(f"{self.a * self.b * self.c}")
        else:
            print(f"{self.a * self.b}")

print(calc.multiply(2, 3, 4))
print(calc.multiply(2, 3))
```

```python
class shape:
    def draw(self):
        print("Drawing shape")

class circle(shape):
    def draw(self):
        super().draw()
        print("Drawing circle")

a=shape()
a.draw()
b=circle()
b.draw()
```

```python
class Shape:
    def draw(self):
        print("Drawing shape")

class Circle(Shape):
    def draw(self):
        # We "override" the parent by not calling super().draw()
        print("Drawing circle")

# --- Execution ---
s = Shape()
s.draw()  # Output: Drawing shape

c = Circle()
c.draw()  # Output: Drawing circle
```

```python
from abc import ABC, abstractmethod

# 1. Define the Abstract Class (The Blueprint)
class Employee(ABC):
    @abstractmethod
    def calculate_salary(self):
        pass

# 2. Define the Subclass (The Implementation)
class Manager(Employee):
    def __init__(self, hours_worked, rate_per_hour):
        self.hours_worked = hours_worked
        self.rate_per_hour = rate_per_hour

    # If you forget to include this or misspell it, Python will throw an error!
    def calculate_salary(self):
        salary = self.hours_worked * self.rate_per_hour
        print(f"Manager Salary: {salary}")

# --- Testing ---
m1 = Manager(1_600, 50)
m1.calculate_salary()
```
