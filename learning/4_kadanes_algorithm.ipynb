### Kadane's Algorithm

```python
# Maximum Subarray sum (Brute force mtd)

def sub_sum(arr):
    n = len(arr)
    maxi = float('-inf')

    for i in range(n):
        total = 0
        for j in range(i, n):
            total = total + arr[j]
            # maxi = max(maxi, total)          <--  Both are ok
            if total > maxi:                   #  <--
                maxi = total

    return maxi

arr = [-2,1,-3,4,-1,2,1,-5,4]
sub_sum(arr)
```

```python
# Maximum Subarray sum (Brute force mtd) and [ finding that sub array ]

def sub_sum(arr):
    n = len(arr)
    maxi = float('-inf')
    start, end = 0, 0

    for i in range(n):
        total = 0
        for j in range(i, n):
            total = total + arr[j]
            if total > maxi:
                maxi = total
                start, end = i, j

    return arr[start : end + 1], maxi
    
arr = [-2,1,-3,4,-1,2,1,-5,4]
res_arr = sub_sum(arr)
res_arr
```

```python
# Maximum Subarray sum (Kadane's Algorithm)

def sum_sub_kadane(arr):
    n = len(arr)
    maxi = float('-inf')
    total = 0

    for i in range(n):
        total = total + arr[i]
        maxi = max(maxi, total)
        if total < 0:
            total = 0

    return maxi

arr = [-2,1,-3,4,-1,2,1,-5,4]
sum_sub_kadane(arr)
```

```python
def kadanes_with_indices(arr):
    if not arr:
        return 0, []

    max_so_far = float('-inf')
    current_max = 0
    start = 0
    end = 0
    temp_start = 0

    for i in range(len(arr)):
        current_max += arr[i]

        if max_so_far < current_max:
            max_so_far = current_max
            start = temp_start
            end = i

        if current_max < 0:
            current_max = 0
            temp_start = i + 1

    return max_so_far, arr[start:end+1]

# Example Usage
arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
max_sum, subarray = kadanes_with_indices(arr)
print(f"Max Sum: {max_sum}, Subarray: {subarray}")
```

```python
def max_subarray_dp(nums):
    """
    Kadane's Algorithm using explicit Dynamic Programming logic.
    Works for all cases, including all-negative arrays.
    """
    if not nums:
        return 0
    
    # Initialize both with the first element
    current_max = max_so_far = nums[0]
    
    for i in range(1, len(nums)):
        # Decide: Start a new subarray at current element, 
        # or extend the existing one?
        current_max = max(nums[i], current_max + nums[i])
        
        # Update the global maximum
        max_so_far = max(max_so_far, current_max)
        
    return max_so_far

# Test Case: All negatives
# Standard Kadane's might return 0; this correctly returns -1
print(f"Max sum (all negatives): {max_subarray_dp([-2, -3, -1, -5])}") 

# Test Case: Mixed numbers
print(f"Max sum (mixed): {max_subarray_dp([-2, 1, -3, 4, -1, 2, 1, -5, 4])}")
```
