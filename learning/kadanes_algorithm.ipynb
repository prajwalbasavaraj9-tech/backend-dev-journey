```python
# Maximum Subarray sum (Brute force mtd)

def sub_sum(arr):
    n = len(arr)
    maxi = float('-inf')

    for i in range(n):
        total = 0
        for j in range(i, n):
            total = total + arr[j]
            # maxi = max(maxi, total)          <--  Both are ok
            if total > maxi:                   #  <--
                maxi = total

    return maxi

arr = [-2,1,-3,4,-1,2,1,-5,4]
sub_sum(arr)
```

```python
# Maximum Subarray sum (Brute force mtd) and [ finding that sub array ]

def sub_sum(arr):
    n = len(arr)
    maxi = float('-inf')
    start, end = 0, 0

    for i in range(n):
        total = 0
        for j in range(i, n):
            total = total + arr[j]
            if total > maxi:
                maxi = total
                start, end = i, j

    return arr[start : end + 1], maxi
    
arr = [-2,1,-3,4,-1,2,1,-5,4]
res_arr = sub_sum(arr)
res_arr
```

```python
# Maximum Subarray sum (Kadane's Algorithm)

def sum_sub_kadane(arr):
    n = len(arr)
    maxi = float('-inf')
    total = 0

    for i in range(n):
        total = total + arr[i]
        maxi = max(maxi, total)
        if total < 0:
            total = 0

    return maxi

arr = [-2,1,-3,4,-1,2,1,-5,4]
sum_sub_kadane(arr)
```
